```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
     
subject = pd.read_csv('.csv', encoding='cp949')
corr_column_names = ['english','limit_num','std_num','reply_num','grade','lecture_ratio','participate_ratio','en_ratio','n_cal_grade','std_num(except W,F,U)','grade_total','num_question','lec_ratio_total','part_ratio_total','std/limit','reply/std','load']
for i in corr_column_names:
    for j in range(len(subject[i])):
        subject[i][j] = float(str(subject[i][j]).replace(",",""))
    subject[i] = subject[i]/max(subject[i])

corr = subject.corr(method='pearson')
print(corr)

fig, ax = plt.subplots(figsize=(18,6))
sns.set(font_scale=1) #    

sns.heatmap(corr, #   
            annot = True, #   
            ax = ax, #   
            linewidths = 0.4, #  
            linecolor = 'white', #  
            fmt = '.1f',
            yticklabels=corr_column_names,
            xticklabels=corr_column_names,           #   
            cmap = 'YlOrRd') # colormap 
             # y  
           # x  
plt.show()
```
